```python
import numpy as np 
import flet as ft
```

```python
w1 = 1000
w2 = 2000
w3 = 500
w4 = 300
selected_location = 0

#selected place
for a in range(10):
    selected_location += a 
    
def main(page: ft.Page):
    page.title = "House Rental Forecast"
    page.window_width = 750
    page.window_height =750
    page.horizontal_alignment = ft.CrossAxisAlignment.CENTER
    rent_text = ft.Text()
    user_name = ""  # Initialize user_name variable
    budget = ""     # Initialize budget variable
    
    #enter click for alertDialog
    def enter_click(e):
        nonlocal user_name, budget
        if not user_name.value:
            user_name.error_text = "Name cannot be blank!"
            user_name.update()
        elif not budget.value:
            budget.error_text = "Budget cannot be blank!"
            budget.update()
        else:
            user_name = user_name.value
            budget = budget.value
            page.session.set("user_name", user_name)
            page.session.set("budget", budget)
            page.dialog.open = False
            page.update()
                    
    #alertDialog for name at begining
    user_name = ft.TextField(label="Enter your name")
    budget = ft.TextField(label="Enter your Budget")
    enter_button = ft.ElevatedButton(text="Enter", on_click=enter_click)
    page.dialog = ft.AlertDialog(
        open=True,
        modal=True,
        title=ft.Text("Welcome!"),
        content=ft.Column([user_name, budget], tight=True),
        actions=[enter_button],
        actions_alignment="end"
    )
    
    #calculate equation
    def calculate_click(area, room, bathroom, location):
        area = float(area)
        room = float(room)
        bathroom = float(bathroom)
        rent = w1 * area + w2 * room + w3 * bathroom + w4 * int(selected_location)
        #display after submited
        rent_text.value = f"Name:{user_name} , Budget: NT{budget}, Calculated Rent: NT{rent} "
        page.update()
    
    #button
    def button_clicked(e):
        global selected_location
        selected_location = l.value
        page.remove(l, b)
        page.add(area, room, bathroom, submit_button, rent_text)
        page.update()
    
    #reset 
    def reset_click(e):
        area.value = ""
        room.value = ""
        bathroom.value = ""
        rent_text.value = ""
        page.update()
        
    #set selection of location       
    l = ft.RadioGroup(content=ft.Column([
        ft.Radio(value=5, label="Taipei"),
        ft.Radio(value=4, label="Xinpei"),
        ft.Radio(value=3, label="Taizhong"),
        ft.Radio(value=2, label="Tainan"),
        ft.Radio(value=1, label="Gaoxiong")]))
    
    #Next button 
    b = ft.ElevatedButton(text='Next', on_click=button_clicked)
    #reset button
    reset_button = ft.ElevatedButton(text="Reset user input", on_click=reset_click)
    page.add(l,b)    
        
    area = ft.TextField(label="Enter the area",hint_text="(坪數) ", width=300)
    room = ft.TextField(label="Enter numbers of the room",hint_text="(間) ", width=300)
    bathroom = ft.TextField(label="Enter numbers of the bathroom",hint_text="(間)", width=300)
    
    def submit_clicked(e):
        calculate_click(area.value, room.value, bathroom.value, l.value)
        rent_text.value = f"Name: {user_name}, Budget: NT{budget},Area: {area.value}, Room: {room.value}, Bathroom: {bathroom.value}, Location: {l.value}"
     
    submit_button = ft.ElevatedButton(text="Submit", on_click=submit_clicked)
    
    page.add(area, room, bathroom, submit_button, reset_button,rent_text)
     
ft.app(target=main)
```
